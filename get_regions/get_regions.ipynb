```python
import pandas as pd 
import numpy as np 
import plotly.express as px
import re
```

```python
# https://www.cbs.nl/nl-nl/maatwerk/2024/35/kerncijfers-wijken-en-buurten-2024
df = pd.read_excel('./kwb-2024.xlsx')
df.shape
```

```python
df.head()
df = df[df['recs'] != "Land"]
```

```python
df.columns
```

```python
inw = 12000
won = round(inw / 1.75)
print(inw, won)
```

```python
df = df[df['a_inw'] >= inw]
df = df[df['a_woning'].astype(int) >= won]
df.shape
```

```python
df['recs'].value_counts()
```

```python
df1 = df[df['recs'] != "Buurt"]
df1.shape
```

```python
df1.head(2)
```

```python
regio_list = df1['regio'].to_list()
len(regio_list)
```

```python
# Filter out items containing "weird characters"
filtered_list = [item for item in regio_list if re.match(r'^[a-zA-Z0-9 ,.-]+$', item)]
# Sort the filtered list alphabetically
filtered_list.sort()
len(filtered_list)
```

```python
filtered_list
```

```python
# Open a file for writing and override existing content
with open('regio_list.txt', 'w', encoding='utf-8') as file:
    # Format the list with 8 items per line
    formatted_list = ',\n'.join(
        ', '.join(f'"{item}"' for item in filtered_list[i:i + 8])
        for i in range(0, len(filtered_list), 8)
    )
    file.write(f"[{formatted_list}]\n")
```
